```python
import numpy as np
import os
import matplotlib.pyplot as plt

# 数据路径
data_dir = '../data/selected_data/SNR_6/'
data_path = os.path.join(data_dir, 'X_SNR_6_mod_128APSK.npy')

# 加载数据
data = np.load(data_path)
print(f"数据形状: {data.shape}")  # 输出: (2, 1024, 2)

# 检查数据维度
num_samples, seq_length, num_features = data.shape
print(f"样本数量: {num_samples}, 序列长度: {seq_length}, 特征数量: {num_features}")

# 可视化设置
plt.figure(figsize=(15, 10))

for sample_idx in range(num_samples):
    plt.subplot(num_samples, 1, sample_idx + 1)
    for feature_idx in range(num_features):
        plt.plot(data[sample_idx, :, feature_idx], label=f'Feature {feature_idx + 1}')
    plt.title(f'Sample {sample_idx + 1}')
    plt.xlabel('Sequence Step')
    plt.ylabel('Feature Value')
    plt.legend()
    plt.grid(True)

plt.tight_layout()
plt.show()
```
